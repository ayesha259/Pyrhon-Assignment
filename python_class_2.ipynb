Create a Program that:
      Ask user for two numbers using input().
      Convert them to float.
      Find their sum.
      Print the type of the result as well.

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
result = num1 + num2
print("The sum is:", result)
print("The type of the result is:", type(result))
```

Write a program for simple calculator:

Write a program for shopping discount calculator:
      Takes the original price of an item.
      Takes the discount percentage.
      Calculates and shows the final price after applying the discount

```python
original_price = float(input("Enter the original price of the item: "))
discount_percent = float(input("Enter the discount percentage: "))
discount_amount = (discount_percent / 100) * original_price
final_price = original_price - discount_amount
print("The final price after discount is:", final_price)
```
