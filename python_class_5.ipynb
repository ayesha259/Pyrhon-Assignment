Calculate and print the sum of even numbers using a for loop from 1 to 50

```python
total = 0
for number in range(1, 51):
    if number % 2 == 0:
        total += number

print("Sum of even numbers from 1 to 50 is:", total)
```

Ask the user for a password in a while loop. Allow 3 tries, and if the correct password is
entered, print "Access granted", otherwise print "Blocked"

```python
correct_password = "1234"
attempts = 0

while attempts < 3:
    password = input("Enter password: ")

    if password == correct_password:
        print("Access granted")
        break
    else:
        attempts += 1
        print("Wrong password")

if attempts == 3 and password != correct_password:
    print("Blocked")
```

Create a countdown from 10 to 1 using a while loop. Use break if it reaches 5 and print
"Midpoint reached"

```python
number = 10
while number >= 1:
    print(number)
    if number == 5:
        print("Midpoint reached")
        break

    number -= 1
```

From 1 to 500, use a loop to find the first number divisible by both 13 and 7. Use break

```python
for number in range(1, 501):
    if number % 13 == 0 and number % 7 == 0:
        print("First number divisible by both 13 and 7 is:", number)
        break
```

From a list of words, create a dictionary where key is the word and value is the length, but
only for words longer than 4 letters.

```python
words = ["apple", "cheery", "banana", "horse", "tiger", "moon"]

word_lengths = {}

for word in words:
    if len(word) > 4:
        word_lengths[word] = len(word)

print(word_lengths)
```

Given a string, build a dictionary using comprehension where the keys are characters and
values are their frequency in the string.

```python
text = "Learning Python"
char_frequency = {char: text.count(char) for char in text}
print(char_frequency)
```

Create a list from 100 to 150 using list comprehension where prime numbers stay as-is, and
non-primes become "Not Prime".


```python
result = []

for num in range(100, 151):
    is_prime = True

    if num < 2:
        is_prime = False
    else:
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break

    if is_prime:
        result.append(num)
    else:
        result.append("Not Prime")

print(result)
```

List Categorizer
Given a list of integers:
l = list(range(1,21))
Separate them into 3 lists:
Even
Odd
Multiples of 5
Print all 3 lists at the end


```python
l = list(range(1, 21))
even = []
odd = []
multiples_of_5 = []
for num in l:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)

    if num % 5 == 0:
        multiples_of_5.append(num)

print("Even:", even)
print("Odd:", odd)
print("Multiples of 5:", multiples_of_5)
```

FizzBuzz Dictionatry
Make a dictionary for numbers 1 to 30 with:
"Fizz" if divisible by 3
"Buzz" if divisible by 5
"FizzBuzz" if divisible by both
"None" otherwise

```python
fizzbuzz_dict = {
    num: ("FizzBuzz" if num % 3 == 0 and num % 5 == 0
          else "Fizz" if num % 3 == 0
          else "Buzz" if num % 5 == 0
          else "None")
    for num in range(1, 31)
}

print(fizzbuzz_dict)
```
