```python
ticket_prices = [250, 300, 200, 400, 150]
print(max(ticket_prices))
print(min(ticket_prices))
ticket_prices.sort(reverse=True)
print(ticket_prices)
ticket_prices.append(350)
print(ticket_prices)
ticket_prices.sort(reverse=True)
print(ticket_prices[:3])
```

Product Prices:
      You have prices of 6 products in your shop.
            Create a list of product prices.
            Increase the price of two products due to tax (1% tax /product).
            Remove the lowest-priced product.
            Calculate the total price of the remaining products

```python
prices = [100, 200, 150, 250, 300, 180]
print(prices)
prices[0] *= 1.01
prices[1] *= 1.01
print(prices)
prices.remove(min(prices))
print(prices)
total_price = sum(prices)
print(total_price)
```

employee payroll record : pyroll system keeps a tuple :(employee_id, name , base_salary, month, net_pay) .
design data for multiple employees and print the one with the highest salary for a given month.
 note : atleast 3 employees records must be created

```python
employees = (
    (101, "Ali", 50000, "January", 48000),
    (102, "Danish", 55000, "January", 52000),
    (103, "Sara", 60000, "February", 58000),
    (104, "Zara", 53000, "January", 51000)
)

search_month = input("Enter month to check highest salary: ")

highest_paid = None

for emp in employees:
    if emp[3] == search_month:
        if highest_paid is None or emp[4] > highest_paid[4]:
            highest_paid = emp

if highest_paid:
    print("\nEmployee with highest salary in", search_month)
    print("ID:", highest_paid[0])
    print("Name:", highest_paid[1])
    print("Net Pay:", highest_paid[4])
else:
    print("No payroll records found for this month.")
```
