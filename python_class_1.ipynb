We have four variables a = 5, b = 10, c = 3 and d = 6 swap the values of variable a & b and c
& d.

```python
a = 5
b = 10
c = 3
d = 6
print("Before swapping:")
print("a =", a, "b =", b)
print("c =", c, "d =", d)
a, b = b, a
c, d = d, c
print("After swapping:")
print("a =", a, "b =", b)
print("c =", c, "d =", d)


```
