      Every time a user visits, their user_id is logged — but users can visit multiple times.
      visits = [101, 102, 103, 101, 104, 102, 105, 103, 105, 106]
      Tasks
      - Find how many unique visitors visited the site.
      - Display the list of unique user_ids (in any order).
      - Compare the total visits vs. unique visitors.
Count Distinct Words in a Sentence:
       - A user enters a sentence (as a single string).
       - Your task is to count how many distinct words are in the sentence.

```python
visits = [101, 102, 103, 101, 104, 102, 105, 103, 105, 106]

unique_users = set(visits)

unique_count = len(unique_users)

total_visits = len(visits)

print("Total visits:", total_visits)
print("Unique visitors:", unique_count)
print("Unique user IDs:", unique_users)
```

```python
sentence = input("Enter a sentence: ")

words = sentence.lower().split()
unique_words = set(words)

print("Number of distinct words:", len(unique_words))
print("Distinct words:", unique_words)
```

Student Performance Tracker using Dictionaries:
Create a dictionary where:
Keys are student names.
Values are lists of their marks in 3 subjects.
Write a code to:
Add three new students with their marks.
Calculate average marks for each student.
Then add student(s) with the average marks into the dictionary.
Convert the dictionary into a list of tuples and sort it by average marks

```python
students = {
    "Ali": [80, 75, 90],
    "Sara": [85, 88, 92]
}

students["Ahmed"] = [70, 65, 75]
students["Ayesha"] = [90, 95, 93]
students["Bilal"] = [60, 70, 65]

averages = {}

for name, marks in students.items():
    avg = sum(marks) / 3
    averages[name] = avg

for name, avg in averages.items():
    students[name] = students[name] + [avg]

student_list = list(averages.items())
student_list.sort(key=lambda x: x[1])

print(students)
print(student_list)
```

Shopping Cart Snapshot:
Instructions:
Create 3 products.
Each product is a dictionary with:
"info" →
 a tuple: (name, category)
"price" →
 integer
"ratings" →
 list of 3 numbers
"tags" →
 set of 3 tags
Store all products in a list called cart.
Questions:
1.What is the category of the second product?
2.What is the total price of all products?
3.Find the average rating of the first product.
4.Combine all tags from the products (using |
operator).
5.Add a new product product4 with your own
data and update cart

```python
product1 = {"info": ("Laptop", "Electronics"), "price": 1200, "ratings": [4.5, 4.0, 5.0], "tags": {"tech", "computer", "portable"}}
product2 = {"info": ("Sneakers", "Footwear"), "price": 150, "ratings": [4.0, 3.5, 4.5], "tags": {"shoes", "sport", "fashion"}}
product3 = {"info": ("Coffee Maker", "Appliances"), "price": 80, "ratings": [4.2, 4.5, 4.0], "tags": {"kitchen", "beverage", "morning"}}
cart = [product1, product2, product3]

second_category = cart[1]["info"][1]
total_price = sum(p["price"] for p in cart)
average_rating = sum(cart[0]["ratings"])/len(cart[0]["ratings"])
all_tags = cart[0]["tags"] | cart[1]["tags"] | cart[2]["tags"]

product4 = {"info": ("Backpack", "Accessories"), "price": 70, "ratings": [4.0, 4.2, 3.8], "tags": {"travel", "bag", "fashion"}}
cart.append(product4)

print(second_category)
print(total_price)
print(average_rating)
print(all_tags)
print(len(cart))
```
